```python
import csv
import pandas as pd

filename = '/content/clearner_dataset.csv'
clean_rows = []

# Open the CSV file and handle the problematic rows
with open(filename, 'r', encoding='utf-8', errors='ignore') as file:
    reader = csv.reader(file)
    for i, row in enumerate(reader):
        if i < 24400:
            continue
        clean_rows.append(row)
df = pd.DataFrame(clean_rows)


print("Data successfully loaded and processed.")
print(df.columns)
```

```python
print(df.columns)
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
df[df.columns[1]] = label_encoder.fit_transform(df[df.columns[1]])
df[df.columns[2]] = label_encoder.fit_transform(df[df.columns[2]])
print(df)
```

```python
df[df.columns[5]] = pd.to_datetime(df[df.columns[5]])
date_mapping = {date: i for i, date in enumerate(sorted(df[df.columns[5]].dt.date.unique()))}
df['Encoded_date'] = df[df.columns[5]].dt.date.map(date_mapping)
print(df)
print(date_mapping)
```
